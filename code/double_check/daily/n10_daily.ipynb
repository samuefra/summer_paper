```python
def read_n10(factor):
    n10_path = f'../../../output/data/double_check/{factor}/n10.parquet'
    df = pd.read_parquet(n10_path)
    return(df)
```

```python
def read_n10_daily(factor):
    n10_path = f'../../../output/data/double_check/{factor}/n10_daily.parquet'
    df = pd.read_parquet(n10_path)
    return(df)
```

### Difference Rate

```python
def rate(df1, df2):
    df = ((df1.reset_index(drop=True)/df2.reset_index(drop=True)) - 1).abs()
    df.index = df1.index
    return(df)
```

```python
rate(size, size_daily)
```

```python
rate(size, size_daily).values.sum()
```

```python
rate(value, value_daily)
```

```python
rate(value, value_daily).values.sum()
```
