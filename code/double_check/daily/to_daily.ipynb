```python
# reading the tickers from .mat file
data = io.loadmat('C:/Users/Usuario/Dropbox/TAQ_Returns/Time_series_tickers.mat')
```

```python
# number of tickers
n_tickers = len(data['All_tickers'])
```

```python
# list that will storage the tickers
all_tickers_timeseries = []

# appending the tickers
for index in range(n_tickers):
    all_tickers_timeseries.append(data['All_tickers'][index][0][0])
```

```python
# this thicker wasn't in the matlab file
all_tickers_timeseries.append('BATS')
```

```python
all_tickers_timeseries
```

```python
def read_n10(factor):
    n10_path = f'../../../output/data/double_check/{factor}/n10.parquet'
    df = pd.read_parquet(n10_path)
    df = df.dropna(axis=0, how='all')
    return(df)
```

```python
size = read_n10('size')
```

```python
daterange = size.index
daterange 
```

```python
df = pd.DataFrame(index=daterange, columns=all_tickers_timeseries)
```

```python
df
```

```python
for date in daterange:
    # setting day in .parquet file name format
    day = str(date)[:4] + str(date)[5:7] + str(date)[8:10]

    # reading the returns dataset
    path = f'../../../input/returns/{day}.parquet'
    returns = pd.read_parquet(path)

    # computing the daily returns
    daily_returns = returns.sum(axis=0)

    # filling the dataframe with daily returns for each firm
    for ticker in daily_returns.index:
        df[ticker][date] = daily_returns[ticker]
```

```python
df
```

```python
# dropping all firms with the whole column filled with NaN
df = df.dropna(axis=1, how='all')
```

```python
output_path = '../../../input/returns/daily.parquet'
df.to_parquet(output_path)
```
