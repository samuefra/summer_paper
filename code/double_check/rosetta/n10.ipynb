## Kozak Factors

```python
daterange = size.index
```

```python
def read_n10_kozak(factor):
    kozak_path = f'../../../input/kozak/daily/n10_{factor}.csv'
    kozak = pd.read_csv(kozak_path, index_col=0)
    kozak.index = pd.to_datetime(kozak.index)
    kozak = kozak.loc[daterange]
    return(kozak)
```

```python
kozak_size = read_n10_kozak('size')
kozak_size
```

```python
kozak_value = read_n10_kozak('value')
kozak_value
```

### Difference Rate

```python
def rate(df1, df2):
    df = ((df1.reset_index(drop=True)/df2.reset_index(drop=True)) - 1).abs()
    df.index = df1.index
    return(df)
```

```python
rate(kozak_size, size_rosetta)
```

```python
rate(size, size_rosetta).values.sum()
```

```python
rate(kozak_value, value_rosetta)
```

```python
rate(value, value_rosetta).values.sum()
```
